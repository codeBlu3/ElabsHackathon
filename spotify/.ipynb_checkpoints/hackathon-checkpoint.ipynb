```python
df = df.dropna() # no day, no month
```

```python
df.head()
```

```python
df.release_year = df.release_year.astype(int)
df.release_month = df.release_month.astype(int)
df.release_day = df.release_day.astype(int)
```

```python
df.dtypes
```

```python
def todate(y,m,d):
    return pd.Timestamp(datetime.datetime(y, m , d))
```

```python
df['rel_date'] = df.apply(lambda x: todate(x['release_year'],x['release_month'], x['release_day']),axis=1)
```

```python
df.rel_date
```

```python
ft5 = lambda df:df[df.position <=5]
```

```python
t5=ft5(df)
```

```python
t5.columns
```

```python
gft5art = t5.groupby(['artist','track_name','track_id'])
```

```python
prel = gft5art['streams'].agg(['count','sum']).reset_index()
```

```python
toptracks =  prel['track_id']
```

```python
toptracks.nunique()
```

```python
ftTtrk = lambda df:df[df.track_id.isin(toptracks)]
```

```python
top5ftTracks = ftTtrk(df)
```

```python
top5ftTracks.head()
```

```python
top5ftTracks.position.max()
```

```python
top5ftTracks['boost_time'] = top5ftTracks.index - top5ftTracks.rel_date
```

```python
top5ftTracks.head()
```

```python
top5ftTracks['cur_date'] = top5ftTracks.index
```

```python
a = top5ftTracks.groupby(['artist','track_name'])['position'].agg([min]).reset_index()
```

```python
boostfilter = top5ftTracks.groupby(['artist','track_name'])['position'].agg([min]).reset_index()
```
